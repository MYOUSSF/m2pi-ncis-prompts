```python
import yake
```

```python
import pandas as pd
from collections import Counter
import nltk
```

```python
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
```

```python
prompts = pd.read_csv('PIMS_Sample_Prompts.csv')
```

```python
prompts
```

```python
language = 'en'
max_ngram = 2
dup_threshold = 0.9
num_words = 20
```

```python
# Initialize YAKE keyword extractor
extractor = yake.KeywordExtractor(lan=language, n=max_ngram, dedupLim=dup_threshold, top=num_words, features=None)
```

```python
# List to store all keywords
all_keywords = []
```

```python
# Extract keywords from each prompt
for prompt in prompts['Prompt'].values:
    for word in extractor.extract_keywords(prompt):
        all_keywords.append(word[0])
```

```python
all_keywords
```

```python
# Count the frequency of each keyword
keyword_counts = Counter(all_keywords)
```

```python
# Sort the keywords by frequency (most common first)
sorted_keywords = keyword_counts.most_common()
```

```python
# Display the sorted keywords with their counts
for keyword, count in sorted_keywords:
    print(f'{keyword}: {count}')
```

```python
sorted_keywords_df = pd.DataFrame(sorted_keywords, columns=['Keyword', 'Count'])
sorted_keywords_df.to_csv('C:\\Users\\ASUS\\Desktop\\sorted_keywords')
```
