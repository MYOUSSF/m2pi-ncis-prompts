```python
import pandas as pd
import yake

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import DBSCAN
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
```

```python
prompts = pd.read_csv('sample-prompts.csv')
prompts
```

```python
language = 'en'
max_ngram = 2
dup_threshold = 0.9
num_words = 20
extractor = yake.KeywordExtractor(lan=language, n=max_ngram, dedupLim=dup_threshold, top=num_words, features=None)
```

```python
prompt_kw = set()
```

```python
for prompt in prompts['Prompt'].values:
    for word in extractor.extract_keywords(prompt):
        prompt_kw.add(word[0])

prompt_kw
```
